```python


# Exercice de compréhension des variables en Python
# 1. Déclarez une variable nommée 'age' et affectez-lui la valeur 25
age = 25

# 2. Créez une variable 'nom' avec votre prénom
nom = "David"

# 3. Affectez la valeur 3.14 à une variable appelée 'pi'
pi = 3.14

# 4. Déclarez une variable booléenne 'est_etudiant' avec la valeur True
est_etudiant = True

print("=== Test de compréhension des variables en Python ===")

# Couleurs et icônes
GREEN_CHECK = "✅"
RED_CROSS = "❌"

# Exercice 1: Déclaration et affectation de variables
print("\nExercice 1: Déclaration et affectation de variables")
try:
   # age = 25
    assert age == 25
    print(f"{GREEN_CHECK} Test 1 réussi - Déclaration 'age'")
except:
    print(f"{RED_CROSS} Échec test 1 - 'age' devrait valoir 25")

try:
    nom = "Pierre"
    assert isinstance(nom, str)
    print(f"{GREEN_CHECK} Test 2 réussi - Déclaration 'nom'")
except:
    print(f"{RED_CROSS} Échec test 2 - 'nom' devrait être une chaîne")

try:
    pi = 3.14
    assert pi == 3.14
    print(f"{GREEN_CHECK} Test 3 réussi - Déclaration 'pi'")
except:
    print(f"{RED_CROSS} Échec test 3 - 'pi' devrait valoir 3.14")

try:
    est_etudiant = True
    assert est_etudiant is True
    print(f"{GREEN_CHECK} Test 4 réussi - Déclaration booléenne")
except:
    print(f"{RED_CROSS} Échec test 4 - 'est_etudiant' devrait être True")

# Exercice 2: Types de variables
print("\nExercice 2: Types de variables")
tests_types = [
    (42, int, "entier"),
    ("Python", str, "chaîne"),
    (3.14159, float, "float"),
    (True, bool, "booléen"),
    ([1, 2, 3], list, "liste")
]

for i, (valeur, type_attendu, desc) in enumerate(tests_types, 1):
    try:
        assert isinstance(valeur, type_attendu)
        print(f"{GREEN_CHECK} Test {i} réussi - Type {desc}")
    except:
        print(f"{RED_CROSS} Échec test {i} - Le type devrait être {type_attendu.__name__}")

# Exercice 3: Opérations avec variables
print("\nExercice 3: Opérations avec variables")
x, y = 10, 3
operations = [
    (x + y, 13, "Addition (10 + 3 = 13)"),
    (x - y, 7, "Soustraction (10 - 3 = 7)"),
    (round(x / y, 2), 3.33, "Division (10 / 3 ≈ 3.33)"),
    (x // y, 3, "Division entière (10 // 3 = 3)"),
    (x % y, 1, "Modulo (10 % 3 = 1)"),
    (x ** y, 1000, "Puissance (10 ** 3 = 1000)")
]

for i, (res, attendu, desc) in enumerate(operations, 1):
    try:
        assert res == attendu
        print(f"{GREEN_CHECK} Test {i} réussi - {desc}")
    except:
        print(f"{RED_CROSS} Échec test {i} - {desc} (obtenu: {res})")

# Exercice 4: Concaténation et conversion
print("\nExercice 4: Concaténation et conversion")
try:
    prenom, age = "Alice", 25
    message = f"{prenom} a {age} ans"
    assert message == "Alice a 25 ans"
    print(f"{GREEN_CHECK} Test 1 réussi - Concaténation f-string")
except:
    print(f"{RED_CROSS} Échec concaténation - Devrait être 'Alice a 25 ans'")

try:
    age_str = str(age)
    assert age_str == "25"
    print(f"{GREEN_CHECK} Test 2 réussi - Conversion en chaîne")
except:
    print(f"{RED_CROSS} Échec conversion - 'age' devrait devenir '25'")

# Exercice 5: Réaffectation de variables
print("\nExercice 5: Réaffectation de variables")
try:
    v = 5
    v += 3
    assert v == 8
    print(f"{GREEN_CHECK} Test réussi - Réaffectation (5 + 3 = 8)")
except:
    print(f"{RED_CROSS} Échec réaffectation - 'v' devrait valoir 8")

# Exercice 6: Échange de variables
print("\nExercice 6: Échange de variables")
try:
    a, b = 10, 20
    a, b = b, a
    assert a == 20 and b == 10
    print(f"{GREEN_CHECK} Test réussi - Échange (a=20, b=10)")
except:
    print(f"{RED_CROSS} Échec échange - a devrait être 20 et b devrait être 10")

# Exercice 7: Portée des variables
print("\nExercice 7: Portée des variables")
try:
    variable_globale = 100
    def fonction_test():
        variable_locale = 50
        return variable_globale + variable_locale
    
    assert fonction_test() == 150
    print(f"{GREEN_CHECK} Test 1 réussi - Portée globale")
    
    try:
        variable_locale
        print(f"{RED_CROSS} Échec portée - variable_locale ne devrait pas être accessible")
    except NameError:
        print(f"{GREEN_CHECK} Test 2 réussi - Portée locale respectée")
except:
    print(f"{RED_CROSS} Échec test portée")

# Exercice 8: Noms de variables valides
print("\nExercice 8: Noms de variables valides")
tests_noms = [
    ("ma_variable", True),
    ("_variable", True),
    ("variable2", True),
    ("Total", True),
    ("2fois", False),
    ("def", False),
    ("total$", False),
    ("nombre-de-pommes", False)
]

for i, (nom, valide) in enumerate(tests_noms, 1):
    try:
        exec(f"{nom} = 1")
        if valide:
            print(f"{GREEN_CHECK} Test {i} réussi - '{nom}' est valide")
        else:
            print(f"{RED_CROSS} Test {i} échec - '{nom}' ne devrait pas être valide")
    except:
        if not valide:
            print(f"{GREEN_CHECK} Test {i} réussi - '{nom}' est bien invalide")
        else:
            print(f"{RED_CROSS} Test {i} échec - '{nom}' devrait être valide")

# Exercice 9: Constantes (convention)
print("\nExercice 9: Constantes (convention)")
try:
    MA_CONSTANTE = 100
    assert MA_CONSTANTE == 100
    print(f"{GREEN_CHECK} Test réussi - Convention MAJUSCULES pour constantes")
except:
    print(f"{RED_CROSS} Échec test constante")

# Exercice 10: None
print("\nExercice 10: None")
try:
    var = None
    assert var is None
    print(f"{GREEN_CHECK} Test réussi - None est bien utilisé")
except:
    print(f"{RED_CROSS} Échec test None")

print("\n=== Fin des tests ===")
```
