## Essentiels de Pandas et NumPy

### Pandas

- **Pandas** est une bibliothèque Python pour la manipulation et l’analyse de données tabulaires.
- Les objets principaux sont :
  - **DataFrame** : tableau de données à deux dimensions (lignes et colonnes).
  - **Series** : colonne ou vecteur à une dimension.

**Fonctions clés :**
- `pd.read_csv("fichier.csv")` : lire un fichier CSV.
- `df.head()` : afficher les premières lignes.
- `df.info()` : informations sur le DataFrame.
- `df.describe()` : statistiques descriptives.
- `df["colonne"]` : accéder à une colonne.
- `df.loc[ligne, colonne]` : accès par étiquette.
- `df.iloc[ligne, colonne]` : accès par indice.
- `df.drop()` : supprimer des lignes ou colonnes.
- `df.groupby("colonne")` : regrouper les données.

### NumPy

- **NumPy** est une bibliothèque pour le calcul scientifique et le traitement de tableaux multidimensionnels.
- L’objet principal est :
  - **ndarray** : tableau de nombres à n dimensions.

**Fonctions clés :**
- `np.array([1, 2, 3])` : créer un tableau.
- `np.zeros((2,3))` : tableau de zéros.
- `np.ones((2,3))` : tableau de uns.
- `np.arange(0, 10, 2)` : tableau d’entiers de 0 à 8.
- `np.mean(a)` : moyenne.
- `np.sum(a)` : somme.
- `np.dot(a, b)` : produit matriciel.
- `np.reshape(a, (n, m))` : changer la forme du tableau.

## DataFrame et Series avec Pandas

- **DataFrame** :  
  Tableau de données à deux dimensions (lignes et colonnes), similaire à une feuille Excel ou une table SQL.  
  Chaque colonne peut contenir un type de données différent.

  ```python
  import pandas as pd

  # Création d'un DataFrame
  df = pd.DataFrame({
      "nom": ["Alice", "Bob", "Charlie"],
      "age": [25, 35, 30]
  })
  print(df)
  ```

- **Series** :  
  Objet à une dimension, représentant une seule colonne ou un vecteur.

  ```python
  # Création d'une Series
  s = pd.Series([10, 20, 30])
  print(s)

  # Extraire une colonne d'un DataFrame (cela donne une Series)
  ages = df["age"]
  print(ages)
  ```

- **Résumé** :  
  - Un **DataFrame** est une collection de **Series** alignées sur un même index.
  

## Indexation et sélection conditionnée avec la librairie Pandas

### Indexation

- **Accéder à une colonne** :
  ```python
  df["nom_colonne"]
  ```
- **Accéder à une ligne par son index** :
  ```python
  df.loc[3]      # Par étiquette (index)
  df.iloc[3]     # Par position (entier)
  ```
- **Accéder à une cellule** :
  ```python
  df.loc[3, "nom_colonne"]
  df.iloc[3, 2]
  ```

### Sélection conditionnée

- **Filtrer les lignes selon une condition** :
  ```python
  df[df["age"] > 30]           # Lignes où la colonne 'age' > 30
  df[df["ville"] == "Paris"]   # Lignes où la colonne 'ville' vaut 'Paris'
  ```

- **Combiner plusieurs conditions** :
  ```python
  df[(df["age"] > 30) & (df["ville"] == "Paris")]
  df[(df["age"] < 18) | (df["ville"] == "Lyon")]
  ```

- **Sélectionner certaines colonnes après filtrage** :
  ```python
  df[df["age"] > 30][["nom", "age"]]
  ```

**Exemple :**

```python
import pandas as pd

df = pd.DataFrame({
    "nom": ["Alice", "Bob", "Charlie"],

    "ville": ["Paris", "Lyon", "Paris"]

## Algèbre linéaire : essentiels pour l’analyse de données

L’algèbre linéaire est une branche des mathématiques qui étudie les vecteurs, les matrices et les opérations sur ces objets.  
Elle est fondamentale en analyse de données, machine learning et statistiques.

### Concepts clés

- **Vecteur** :  
  Un vecteur est une liste ordonnée de nombres, par exemple `[2, 5, 7]`.  
  Il peut représenter une observation (ex : les notes d’un élève) ou une variable (ex : les températures sur plusieurs jours).

- **Matrice** :  
  Une matrice est un tableau rectangulaire de nombres (2D array), composé de lignes et de colonnes.  
  Elle permet de stocker plusieurs vecteurs, comme un tableau de données où chaque ligne est une observation et chaque colonne une variable.

- **Opérations sur matrices** :  
  - **Addition** : additionner deux matrices de même taille.
  - **Multiplication** : produit matriciel, utile pour les transformations et les modèles.
  - **Transposition** : inverser lignes et colonnes.
  - **Inversion** : trouver la matrice inverse (utile pour résoudre des systèmes).

- **Système d’équations linéaires** :  
  Résoudre plusieurs équations simultanées, par exemple pour trouver les paramètres d’un modèle.

- **Valeurs propres et vecteurs propres** :  
  Utilisés pour la réduction de dimension (ex : PCA), pour comprendre la structure des données.

- **Norme** :  
  Mesure la “taille” ou la longueur d’un vecteur (distance à l’origine).

### Utilité en analyse de données

- **Stockage des données** :  
  Les données tabulaires sont souvent représentées sous forme de matrices pour faciliter les calculs.

- **Transformation des données** :  
  Les opérations matricielles permettent de manipuler, transformer et combiner les données.

- **Réduction de dimension** :  
  Des techniques comme la PCA (Analyse en Composantes Principales) utilisent l’algèbre linéaire pour simplifier les données et extraire l’essentiel.

- **Modélisation** :  
  Les modèles statistiques et de machine learning utilisent des opérations sur vecteurs et matrices pour l’entraînement et la prédiction.

### Exemple simple en Python avec NumPy

```python
import numpy as np

# Création d’un vecteur
v = np.array([1, 2, 3])

# Création d’une matrice
M = np.array([[1, 2], [3, 4]])

# Addition de matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
addition = A + B

# Multiplication de matrices
resultat = np.dot(M, v[:2])

# Transposition
transposée = M.T

print("Addition :\n", addition)
print("Multiplication :\n", resultat)

````python
import numpy as np
v = np.array([1, 2, 3])

# Création d’une matrice
M = np.array([[1, 2], [3, 4]])

# Addition de matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
addition = A + B

# Multiplication de matrices
resultat = np.dot(A,B)

# Transposition
transposée = M.T
print("all",v[:2])

print("Addition :\n", addition)
print("Multiplication :\n", resultat)
print("Transposée :\n", transposée)
````
